```python
import phylonetwork
import matplotlib
import networkx
import pygraphviz
%matplotlib inline
```

```python
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
def split(net,u):
    utilde = net._generate_new_id()
    for e in net.in_edges(u):
        print(e)
        net.add_edge(e[0],utilde)
        net.remove_edge(e[0],u)
    net.add_edge(utilde,u)
    return utilde
```

```python
def augmentation_T(net,ell,wt,vts):
    netb = net.copy()
    netb.cache = {}
    r = len(vts) + 1
    us = [netb._generate_new_id() for _ in range(r)]
    netb.add_path(us)
    netb._labels[us[-1]]=ell
    w = split(netb,wt)
    netb.add_edge(w,us[0])
    for i in range(r-1):
        u = us[i]
        v = split(netb,vts[i])
        netb.add_edge(u,v)
    if networkx.is_directed_acyclic_graph(netb):
        return netb
    else:
        return None
```

```python
t = phylonetwork.PhyloNetwork(eNewick="(1,2);")
```

```python
tb = augmentation_T(t,'3','_1',[])
```

```python
networkx.draw_networkx(t)
```

```python
pos = graphviz_layout(tb)
networkx.draw_networkx(tb,pos,labels=tb._labels)
```
