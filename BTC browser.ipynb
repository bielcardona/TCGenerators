```python
from generating_TC import *
taxa = [str(i+1) for i in range(5)]
btc = BTC_networks(taxa)
N = 0
nets = []
for n in btc:
    N += 1
    nets.append(n)
print(f"There are {N} BTC networks over {taxa}.")
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Button
%matplotlib notebook

fig, ax = plt.subplots()
xx=fig.get_axes()[0]
plt.subplots_adjust(bottom=0.2)


class Index(object):
    ind = None

    def next(self, event):
        if self.ind is None:
            self.ind = 0
        else:
            self.ind += 1
        i = self.ind % len(nets)
        plt.sca(xx)
        plt.cla()
        draw(nets[i])

    def prev(self, event):
        if self.ind is None:
            self.ind = 0
        else:
            self.ind -= 1
        i = self.ind % len(nets)
        plt.sca(xx)
        plt.cla()
        draw(nets[i])

callback = Index()
axprev = plt.axes([0.7, 0.05, 0.1, 0.075])
axnext = plt.axes([0.81, 0.05, 0.1, 0.075])
bnext = Button(axnext, 'Next')
bnext.on_clicked(callback.next)
bprev = Button(axprev, 'Previous')
bprev.on_clicked(callback.prev)

plt.show()
callback.next(None)
```
